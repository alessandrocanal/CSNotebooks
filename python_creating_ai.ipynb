# Intro
This is the recipe to create an AI
1. Data preparation: data is the raw material for a model to learn and generate significant results
2. Understand the main types of learning (supervised, unsupervised)
3. Understand ML algorithms
4. Evaluate model performance to avoid *overfitting*
5. Learn about Python libraries (Pandas, Numpy, matplotlib, Seaborn, Scikit-learn, Tensorflow, Pytorch)
6. Optimize models by setting **hyperparameters**
7. Identify a practical problem

## Data


* Retrieval
* Exploration and cleaning
* Transformation into numerical format
* Division into training, validation, and test sets
* Management of quantity and quality (augmentation)
* Optimization of processing resources
* Data analysis to make decisions about possible changes or improvements in the collection
